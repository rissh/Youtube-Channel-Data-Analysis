## Analysing Likes

```python
# Some Data preprocessing on Likes

df['likes']   
```

```python
# Likes in dataset are in string, So we have to convert it in Integer 
# Another thing is some likes are in form of 'K' means in terms of 1000. This is also we have to convert into Integer

likes  = []
for i in df['likes']:
    if('K' in i):
        likes.append(int(float(i[:-1])) * 1000)
        
    else:
        likes.append(int(i))
```

```python
# It's time to make changes in original dataset
df['likes'] = likes
df
```

## Analysing Durantion of Video

```python
df['duration']
```

```python
# Time Duration in dataset are in string, So we have to convert it in Integer 
# Format of Time duration is [HH : MM : SS]
# Now we will convert it into Seconds.

duration = []
for i in df['duration']:
    time = i.split(',')
    
    sec = 0
    if len(time) == 1:                        # if Lenght of time is equal to 1
        if(i.split(' ')[1] == 'seconds'):
            sec += int(i.split(' ')[0])
        if(i.split(' ')[1] == 'minutes'):
            sec += (int(i.split(' ')[0]) * 60)
        if(i.split(' ')[1] == 'hour'):
            sec += (int(i.split(' ')[0]) * 3600)
            
        duration.append(sec)    
      
    sec = 0
    if len(time) == 2:                        # if Lenght of time is equal to 2
        if(time[0].split(' ')[1] == 'minutes'):
            sec += int(time[0].split(' ')[0]) * 60
        if(time[0].split(' ')[1] == 'hour'):
            sec += int(time[0].split(' ')[0]) * 3600
        if(time[1].strip().split(' ')[1] == 'seconds'):
            sec += int(time[1].strip().split(' ')[0])
        if(time[1].strip().split(' ')[1] == 'minutes'):
            sec += int(time[1].strip().split(' ')[0]) * 60
            
        duration.append(sec)
    
    sec = 0
    if(len(time) == 3):                        # if Lenght of time is equal to 3
        #print(time)
        if(time[1].strip().split(' ')[1] == 'hour' or 'hours'):
            sec += int(time[1].strip().split(' ')[0]) * 3600
        if(time[1].strip().split(' ')[1] == 'minutes'):
            sec += int(time[1].strip().split(' ')[0]) * 60
        if(time[1].strip().split(' ')[1] == 'seconds'):
            sec += int(time[1].strip().split(' ')[0])
            
        duration.append(sec)       
```

```python
# It's time to make changes in original dataset

df['duration'] = duration
df
```
